```python
from google.colab import files
import pandas as pd
import io

# This will open a file picker
uploaded = files.upload()

# Read the specific file once uploaded
filename = 'customer_shopping_behavior.csv'
if filename in uploaded:
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
    display(df.head())
else:
    print(f'Please upload a file named {filename}')
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Created a column age_group to group customers into age groups
labels=['Young Adult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Created Column purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly': 7,
    'Monthly':30,
    'Quaterly':90,
    'Bi-weekly':14,
    'Annually': 365,
    'Every 3 months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
# Deleteing promo code column since it is redundant with same data in discount applied column
df=df.drop('promo_code_used',axis=1)
```
